```python
import pandas as pd
```

```python
df = pd.read_csv("student_behavior_extended.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df["day_name"] = df["date"].dt.day_name()
df["is_weekend"] = df["day_name"].isin(["Saturday", "Sunday"])
```

```python
df.info()
```

```python
df.rename(columns= {'is_weekend':'weekend_day' } , inplace= True)
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="RdBu")
plt.title("Heatmap")
plt.show()
```

```python
df.groupby("weekend_day")[["study_hours","phone_usage_hours","productivity_score"]].mean()
```

```python
sns.countplot(x='burnout_level' , data = df)
plt.title("Burnout Level Distribution")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x = 'sleep_hours' , y = 'productivity_score', data = df , color="crimson")
plt.title("sleep vs productivity")
plt.grid(True)
plt.show()

plt.figure(figsize=(7,5))
sns.scatterplot(x="phone_usage_hours", y="distractions", data=df , color = "green")
plt.title("phone vs distrcation")
plt.grid(True)
plt.show()

plt.figure(figsize=(7,5))
sns.scatterplot(x="exercise_minutes", y="mood", data=df , color = "blue")
plt.title("exercise vs mood")
plt.grid(True)
plt.show()
```

 see in plot when the sleep increase then productivity  also increases and same but negatively  when the phone usage increases 
 distraction also increase
